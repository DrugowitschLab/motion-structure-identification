```python
from analysis.data_exp1 import DataExp1
from analysis import models
import matplotlib.pyplot as plt
import numpy as np
Π = np.arange(0.02, 0.40, 0.02)
Σ = np.arange(0.1, 2.1, 0.1)
BIC = np.zeros((len(Π), len(Σ)))
baseline = 0
for pid in DataExp1.pids:
    data = DataExp1(pid)
    model = data.build_model(models.ChoiceModel6Param)
    baseline += model.fit().bic
baseline /= len(DataExp1.pids)
for i in range(len(Π)):
    print(i)
    models.ChoiceModel4Param.π_L = Π[i]
    for j in range(len(Σ)):
        models.ChoiceModel4Param.σ_obs = Σ[j]
        for pid in DataExp1.pids:
            data = DataExp1(pid)
            model = data.build_model(models.ChoiceModel4Param)
            BIC[i, j] += model.fit().bic
BIC /= len(DataExp1.pids)
BIC -= baseline
import pickle
with open('../data/exp1/grid.dat', 'wb') as f:
    pickle.dump(BIC, f)
vmax = np.abs(BIC).max()
_, ax = plt.subplots()
im = ax.imshow(BIC, interpolation='nearest', vmin=-vmax, vmax=vmax, cmap=plt.get_cmap('bwr'))
plt.colorbar(im, ax=ax)
ax.set_title('BIC(4-param)-BIC(6-param)')
ax.set_xlabel('σ_obs')
ax.set_ylabel('π_L')
ax.set_xticks(np.arange(len(Σ)))
ax.set_xticklabels(map(lambda f: f'{f:.2f}', Σ), rotation=45)
ax.set_yticks(np.arange(len(Π)))
ax.set_yticklabels(map(lambda f: f'{f:.2f}', Π))
```
