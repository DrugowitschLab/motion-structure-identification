```python
from analysis.data_exp1 import DataExp1
import analysis.models as models
import matplotlib.pyplot as plt
import numpy as np

for pid in DataExp1.pids:
    print(pid)
    data = DataExp1(pid)
    _, ax = plt.subplots(1, 3)
    data.plot_confusion_matrix(ax[0])
    model = data.build_model(models.ChoiceModel4Param)
    res = model.fit()
    model.plot_confusion_matrix(model.predict(res), ax[1])
    print(res.log_likelihood)
    cv_pred = data.cross_validate(models.ChoiceModel4Param)
    model.plot_confusion_matrix(cv_pred, ax[2])
    cv_pred['choice'] = data.df['choice']
    print(np.log(cv_pred.apply(lambda row: row[row['choice']], axis=1)).sum())
    plt.show()
    print()
```
